```python
# Shotmap data
games_fcb = ["11396480", "11396470", "11396456", "11396452", "11396443"]
players_fcb = pd.DataFrame(pd.read_html('https://fbref.com/de/mannschaften/054efa67/Bayern-Munich-Statistiken')[0]).iloc[:,0]

shots_fcb = []
for x in games_fcb:
    shots_fcb.append([z for z in requests.get('https://api.sofascore.com/api/v1/event/' + x + '/shotmap', headers = headers).json()['shotmap'] if (z['player']['name'] in players_fcb.values) == True])

data_fcb = []
for x in shots_fcb:
    data_fcb.append(pd.DataFrame({
                'name': [z['player']['shortName'] for z in x],
                'position': [z['player']['position'] for z in x],
                'shot_type': [z['shotType'] for z in x],
                'xg': [z['xg'] for z in x], 
                'body_part': [z['bodyPart'] for z in x], 
                'goal_location': [z['goalMouthLocation'] for z in x], 
                'time_sec': [z['timeSeconds'] for z in x],
                'shot_loc_x': [z['playerCoordinates']['x'] for z in x],
                'shot_loc_y': [z['playerCoordinates']['y'] for z in x],
                'shot_loc_z': [z['playerCoordinates']['z'] for z in x]
                }))
    
#-----------------------------------------------------------------------------

games_city = ["11352303", "11352389", "11352524", "11352594", "11352625"]
players_city = pd.DataFrame(pd.read_html('https://fbref.com/de/mannschaften/b8fd03ef/Manchester-City-Statistiken')[0]).iloc[:,0]

shots_city = []
for x in games_city:
    shots_city.append([z for z in requests.get('https://api.sofascore.com/api/v1/event/' + x + '/shotmap', headers = headers).json()['shotmap'] if (z['player']['name'] in players_city.values) == True])

data_city = []
for x in shots_city:
    data_city.append(pd.DataFrame({
                'name': [z['player']['shortName'] for z in x],
                'position': [z['player']['position'] for z in x],
                'shot_type': [z['shotType'] for z in x],
                'xg': [z['xg'] for z in x], 
                'body_part': [z['bodyPart'] for z in x], 
                'goal_location': [z['goalMouthLocation'] for z in x], 
                'time_sec': [z['timeSeconds'] for z in x],
                'shot_loc_x': [z['playerCoordinates']['x'] for z in x],
                'shot_loc_y': [z['playerCoordinates']['y'] for z in x],
                'shot_loc_z': [z['playerCoordinates']['z'] for z in x]
                }))
```

```python
name_fcb = pd.DataFrame(pd.concat(data_fcb)['name'].value_counts()).reset_index(names = 'Name').rename(columns = {'name': 'Shots'})
name_fcb['Club'] = 'Bayern München'

name_city = pd.DataFrame(pd.concat(data_city)['name'].value_counts()).reset_index(names = 'Name').rename(columns = {'name': 'Shots'})
name_city['Club'] = 'Manchester City'

name = pd.concat([name_fcb, name_city]).to_csv('data/name.csv')
```

```python
position_fcb = pd.DataFrame(pd.concat(data_fcb)['position'].value_counts()).reset_index(names = 'Position').rename(columns = {'position': 'Shots'}).reindex([2, 0, 1])
position_fcb['Club'] = 'Bayern München'

position_city = pd.DataFrame(pd.concat(data_city)['position'].value_counts()).reset_index(names = 'Position').rename(columns = {'position': 'Shots'}).reindex([2, 1, 0]) 
position_city['Club'] = 'Manchester City'

position = pd.concat([position_fcb, position_city]).replace({'Position': {'D': 'Defender', 'M': 'Midfielder', 'F': 'Forward'}}).to_csv('data/position.csv')
```

```python
shot_type_fcb = pd.DataFrame(pd.concat(data_fcb)['shot_type'].value_counts()).reset_index(names = 'ShotType').rename(columns = {'shot_type': 'Shots'}).reindex([2, 0, 1, 4, 3])
shot_type_fcb['Club'] = 'Bayern München'

shot_type_city = pd.DataFrame(pd.concat(data_city)['shot_type'].value_counts()).reset_index(names = 'ShotType').rename(columns = {'shot_type': 'Shots'}).reindex([2, 0, 1, 4, 3]) 
shot_type_city['Club'] = 'Manchester City'

shot_type = pd.concat([shot_type_fcb, shot_type_city]).to_csv('data/shot_type.csv')
```

```python
body_part_fcb = pd.DataFrame(pd.concat(data_fcb)['body_part'].value_counts()).reset_index(names = 'BodyPart').rename(columns = {'body_part': 'Shots'}) 
body_part_fcb['Club'] = 'Bayern München'

body_part_city = pd.DataFrame(pd.concat(data_city)['body_part'].value_counts()).reset_index(names = 'BodyPart').rename(columns = {'body_part': 'Shots'}).reindex([1, 0, 2]) 
body_part_city['Club'] = 'Manchester City'

body_part = pd.concat([body_part_fcb, body_part_city]).to_csv('data/body_part.csv')
```

```python
goal_location_fcb = pd.DataFrame(pd.concat(data_fcb)['goal_location'].value_counts()).reset_index(names = 'GoalLocation').rename(columns = {'goal_location': 'Shots'}) 
goal_location_fcb['Club'] = 'Bayern München'

goal_location_city = pd.DataFrame(pd.concat(data_city)['goal_location'].value_counts()).reset_index(names = 'GoalLocation').rename(columns = {'goal_location': 'Shots'}) 
goal_location_city['Club'] = 'Manchester City'

goal_location = pd.concat([goal_location_fcb, goal_location_city])
goal_location['What'] = np.select([goal_location['GoalLocation'].str.contains('right'), goal_location['GoalLocation'].str.contains('left'), goal_location['GoalLocation'].str.contains('centre')], ['right', 'left', 'centre'], default = 'high')
goal_location.to_csv('data/goal_location.csv')
```

```python
xg_fcb = pd.DataFrame(pd.concat(data_fcb)['xg']).rename(columns = {'xg': 'Value'})
xg_fcb['Club'] = 'Bayern München'

xg_city = pd.DataFrame(pd.concat(data_city)['xg']).rename(columns = {'xg': 'Value'})
xg_city['Club'] = 'Manchester City'

xg = pd.concat([xg_fcb, xg_city])
xg['What'] = 'xG'

time_fcb = pd.DataFrame(pd.concat(data_fcb)['time_sec']).rename(columns = {'time_sec': 'Value'})
time_fcb['Club'] = 'Bayern München'

time_city = pd.DataFrame(pd.concat(data_city)['time_sec']).rename(columns = {'time_sec': 'Value'})
time_city['Club'] = 'Manchester City'

time = pd.concat([time_fcb, time_city])
time['Value'] = round(time['Value'] / 60, 2)
time['What'] = 'Time [min]'

time_xg = pd.concat([xg, time]).to_csv('data/time_xg.csv')
```

```python
shot_position_fcb =  pd.concat(data_fcb)[['shot_loc_x', 'shot_loc_y']]
shot_position_fcb['Inside'] = np.where((shot_position_fcb['shot_loc_x'] >= 0) & (shot_position_fcb['shot_loc_x'] <= 17) & (shot_position_fcb['shot_loc_y'] > 21) & (shot_position_fcb['shot_loc_y'] < 79), "Inside", "Outside")
shot_position_fcb['Side'] = np.select([shot_position_fcb['shot_loc_y'] >= 66.7, shot_position_fcb['shot_loc_y'] <= 33.3], ['Right', 'Left'], default = 'Middle')
shot_position_fcb = shot_position_fcb.reset_index()
for i in range(len(shot_position_fcb)):
    shot_position_fcb.at[i, 'Distance'] = np.linalg.norm(shot_position_fcb.loc[i, ["shot_loc_x", 'shot_loc_y']]-[0,50])
shot_position_fcb['Club'] = 'Bayern München'

shot_position_city =  pd.concat(data_city)[['shot_loc_x', 'shot_loc_y']]
shot_position_city['Inside'] = np.where((shot_position_city['shot_loc_x'] >= 0) & (shot_position_city['shot_loc_x'] <= 17) & (shot_position_city['shot_loc_y'] > 21) & (shot_position_city['shot_loc_y'] < 79), "Inside", "Outside")
shot_position_city['Side'] = np.select([shot_position_city['shot_loc_y'] >= 66.7, shot_position_city['shot_loc_y'] <= 33.3], ['Right', 'Left'], default = 'Middle')
shot_position_city = shot_position_city.reset_index()
for i in range(len(shot_position_city)):
    shot_position_city.at[i, 'Distance'] = np.linalg.norm(shot_position_city.loc[i, ["shot_loc_x", 'shot_loc_y']]-[0,50])
shot_position_city['Club'] = 'Manchester City'

shot_position = pd.concat([shot_position_fcb, shot_position_city]).to_csv('data/shot_position.csv')
```

```python
inside_fcb = pd.DataFrame(shot_position_fcb[['Inside']].value_counts()).reset_index(names = 'Value')
inside_fcb['What'] = 'Penalty Area'

side_fcb = pd.DataFrame(shot_position_fcb[['Side']].value_counts()).reset_index(names = 'Value').reindex([2, 1, 0])
side_fcb['What'] = 'Side of the Pitch'

insides_fcb = pd.concat([inside_fcb, side_fcb])
insides_fcb['Club'] = 'Bayern München'

inside_city = pd.DataFrame(shot_position_city[['Inside']].value_counts()).reset_index(names = 'Value')
inside_city['What'] = 'Penalty Area'

side_city = pd.DataFrame(shot_position_city[['Side']].value_counts()).reset_index(names = 'Value').reindex([2, 1, 0])
side_city['What'] = 'Side of the Pitch'

insides_city = pd.concat([inside_city, side_city])
insides_city['Club'] = 'Manchester City'

insides = pd.concat([insides_fcb, insides_city])
insides.columns = ['Value', 'Shots', 'What', 'Club']
insides.to_csv('data/insides.csv')
```
